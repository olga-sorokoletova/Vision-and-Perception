#**Main function that calculates gradients through the Laplacian and Horisontal/Vertical/Combined Sobel**#

```python
def gradients(img_path):
    #laplacian gradient
    im1 = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    lap = cv2.Laplacian(im1, cv2.CV_64F, ksize = 1)
    lap = np.uint8(np.absolute(lap))

    #sobel gradient representation
    sobelX = cv2.Sobel(im1, cv2.CV_64F, 1, 0)
    sobelY = cv2.Sobel(im1, cv2.CV_64F, 0, 1)
    sobelX = np.uint8(np.absolute(sobelX))
    sobelY = np.uint8(np.absolute(sobelY))

    sobelCombined = cv2.bitwise_or(sobelX, sobelY)

    fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows = 1, ncols = 5, figsize = (20, 10))
    ax1.imshow(im1, cmap='gray')
    ax1.set_title("Image")
    ax2.imshow(lap, cmap='gray')
    ax2.set_title("Laplacian")
    ax3.imshow(sobelX, cmap='gray')
    ax3.set_title("SobelX")
    ax4.imshow(sobelY, cmap='gray')
    ax4.set_title("SobelY")
    ax5.imshow(sobelCombined, cmap='gray')
    ax5.set_title("Sobel Combined")
    print("")
```
